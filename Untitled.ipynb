```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as plt
import seaborn as sns
import os

df = pd.read_csv("./Influenza_data_BR_2010-2016.csv")
print(df)
```

```python
print(df['A (H1)'])
```

```python
t = df['YEAR']
s1 = df['A (H1)']
s2 = df['A (H1N1)pdm09']
s3 = df['A (H3)']
s4 = df['A (H5)']
s5 = df['A (not subtyped)']
s6 = df['A (Total)']
s7 = df['B (Yamagata lineage)']
s8 = df['B (Victoria lineage)']
s9 = df['B (lineage not determined)']
s10= df['B (Total)']

fig, ax = plt.subplots(figsize=(15,6))
plt.plot(t, s1,'r--', label = 'A H1')
plt.plot(t, s2, 'b--', label = 'A H1N1 pdm09')
plt.plot(t, s3, 'g--', label = 'A H3') 
plt.plot (t, s4, 'y--', label = 'A H5')
plt.plot(t, s5, 'k--', label = 'A not subtyped')
plt.plot(t, s6, 'c--', label = 'A total') 
plt.plot(t, s7, 'k-', label = 'B Yamagata')
plt.plot(t, s8, 'r-', label = 'B Victoria')
plt.plot(t, s9, 'b-', label = 'B undetermined')
plt.plot(t, s10, 'g-', label = 'B total')
plt.legend()
ax.set_title("Influenza in Brazil 2010-2016", fontdict={'fontsize':15})
ax.set_xlabel('Years', fontdict={'fontsize':14})
ax.set_ylabel('Deaths', fontdict={'fontsize':14})

plt.show()
```

```python
t = df['YEAR']

s6 = df['A (Total)']
s10= df['B (Total)']


fig, ax = plt.subplots(figsize=(15,6))
plt.plot(t, s6, 'c-', label = 'A total') 
plt.plot(t, s10, 'g-', label = 'B total')
plt.legend()

ax.set_title("Influenza A and B in Brazil 2010-2016", fontdict={'fontsize':15})
ax.set_xlabel('Years', fontdict={'fontsize':14})
ax.set_ylabel('Deaths', fontdict={'fontsize':14})

plt.show()
```

# A total and B total, till 2015, seem to present similar curve behavior.


```python
t = df['YEAR']

s6 = df['A (Total)']
s10= df['B (Total)']
s3 = df['A (H3)']



fig, ax = plt.subplots(figsize=(15,6))
plt.plot(t, s6, 'c-', label = 'A total') 
plt.plot(t, s10, 'g-', label = 'B total')
plt.plot(t, s3, 'r-', label = 'A H3')
plt.legend()

ax.set_title("Influenza A and B in Brazil 2010-2016", fontdict={'fontsize':15})
ax.set_xlabel('Years', fontdict={'fontsize':14})
ax.set_ylabel('Deaths', fontdict={'fontsize':14})

plt.show()
```

# A H3 curve peaks seem to be mirrored with A/B total curve peaks
